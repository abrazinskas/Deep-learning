## Section 2: Multinomial Logistic Regression [5 points]

In this section you will get started by implementing a linear classification model called [Multinomial Logistic Regression](http://ufldl.stanford.edu/wiki/index.php/Softmax_Regression). Later on you will extend this model to a neural network. You will train it by using the [mini-batch Stochastic Gradient Descent algorithm](http://sebastianruder.com/optimizing-gradient-descent/index.html#minibatchgradientdescent). You should implement how to sample batches, how to compute the loss, how to compute the gradient of the loss with respect to the parameters of the model and how to update the parameters of the model. 

You should get around 0.35 accuracy on the validation and test sets with the provided parameters.


```python
# negative cross entropy
def loss(X, y, W, b, weight_decay):
    m = X.shape[0]
    pred = h(X,W,b)
    loss = - np.mean(np.log(pred[range(m),y]))
    return loss + weight_decay * L2(W)/2.0

# expected loss
def exp_loss(X, y):
    m = X.shape[0]
    pred = np.ones((m, 10)) /10.0
    loss = - np.sum(np.log(pred[range(m),y]))/float(m)
    return loss

# generatilized L2 norm for matrices, assuming that # of rows are # of features
def L2(W):
    vec_norms = np.sum(W**2, axis=0)
    return np.sum(vec_norms)
    
    
# accuracy is based on argmax
def accuracy(X, y, W, b):
    pred = np.argmax(h(X, W, b), axis=1)
    return float(np.sum(pred == y)) / float(len(y))
    

# hypothesis(softmax) function that returns a vector over classes
def h(x, W, b):
    num = np.exp(np.dot(x, W) + b)
    return num / np.sum(num, axis=1).reshape((-1,1))

def grad_loss(X, y, W, b, weight_decay):
    k = W.shape[1] # number of classes
    m = X.shape[0]
    pred = h(X,W,b) # [m x k]
    pred[range(m), y]-= 1 
    # compute gradient
    dW = np.dot(X.T, pred)/float(m) + weight_decay * W 
    db = np.mean(pred, axis=0)
    return dW, db
```

```python
# DONT CHANGE THE SEED AND THE DEFAULT PARAMETERS. OTHERWISE WE WILL NOT BE ABLE TO CORRECT YOUR ASSIGNMENT!
# Seed
np.random.seed(42)

# Default parameters. 
num_iterations = 1500
val_iteration = 100
batch_size = 200
learning_rate = 1e-7

weight_decay = 3e+4
weight_scale = 0.0001

nr_of_feat = X_train.shape[1]
########################################################################################                                                                            #
# Initialize the weights W using a normal distribution with mean = 0 and std =         #
# weight_scale. Initialize the biases b with 0.                                        #
########################################################################################   
W = np.random.normal(size=(nr_of_feat, num_classes),loc =0, scale = weight_scale)
b = np.zeros((num_classes, ),dtype="float32")
########################################################################################
#                              END OF YOUR CODE                                        #
########################################################################################

train_loss_history = []
train_acc_history = []

val_loss_history = []
val_acc_history = []


for iteration in range(num_iterations):
    ########################################################################################
    # Sample a random mini-batch with the size of batch_size from the train set. Put the   #
    # images to X_train_batch and labels to Y_train_batch variables.                       #
    ########################################################################################
    sample = np.random.choice(X_train.shape[0], size= batch_size, replace=False)
    X_train_batch = X_train[sample]
    Y_train_batch = Y_train[sample]
    ########################################################################################
    #                              END OF YOUR CODE                                        #
    ########################################################################################
    
    ########################################################################################
    # Compute the loss and the accuracy of the multinomial logistic regression classifier  #
    # on X_train_batch, Y_train_batch. The loss should be an average of the losses on all  #
    # samples in the mini-batch. Include to the loss L2-regularization over the weight     #
    # matrix W with regularization parameter equals to weight_decay.                       #            
    ########################################################################################
    train_loss = loss(X_train_batch, Y_train_batch, W, b, weight_decay)
    train_acc = accuracy(X_train_batch, Y_train_batch, W, b)
    ########################################################################################
    #                              END OF YOUR CODE                                        #
    ########################################################################################
    
    ########################################################################################
    # Compute the gradients of the loss with the respect to the weights and biases. Put    #
    # them in dW and db variables.                                                         #
    ########################################################################################
    dW, db = grad_loss(X_train_batch, Y_train_batch, W, b, weight_decay)
    ########################################################################################
    #                              END OF YOUR CODE                                        #
    ########################################################################################
    
    ########################################################################################
    # Update the weights W and biases b using the Stochastic Gradient Descent update rule. #
    ########################################################################################
    W -= learning_rate * dW 
    b -= learning_rate * db
    
    ########################################################################################
    #                              END OF YOUR CODE                                        #
    ########################################################################################
    
    if iteration % val_iteration == 0 or iteration == num_iterations - 1:
        ########################################################################################
        # Compute the loss and the accuracy on the validation set.                             #
        ########################################################################################
        val_loss = loss(X_val, Y_val, W, b, weight_decay)
        val_acc = accuracy(X_val, Y_val, W, b)
        ########################################################################################
        #                              END OF YOUR CODE                                        #
        ########################################################################################
        train_loss_history.append(train_loss)
        train_acc_history.append(train_acc)
        val_loss_history.append(val_loss)
        val_acc_history.append(val_acc)
        
        # Output loss and accuracy during training
        print("Iteration {0:d}/{1:d}. Train Loss = {2:.3f}, Train Accuracy = {3:.3f}".
              format(iteration, num_iterations, train_loss, train_acc))
        print("Iteration {0:d}/{1:d}. Validation Loss = {2:.3f}, Validation Accuracy = {3:.3f}".
              format(iteration, num_iterations, val_loss, val_acc))

########################################################################################
# Compute the accuracy on the test set.                                                #
########################################################################################
test_acc = accuracy(X_test, Y_test, W, b)
########################################################################################
#                              END OF YOUR CODE                                        #
########################################################################################
print("Test Accuracy = {0:.3f}".format(test_acc))
```

### Multinomial Logistic Regression: Question 2 [4 points]

Name at least three factors that determine the size of batches in practice and briefly motivate your answers. The factors might be related to computational or performance aspects.

1) The memory size of GPU or RAM (if CPUs are used). I.e. you can process from memory (e.g. compute forward and backward passes) as much data as memory allows. Usually working directly with data stored on the hard disk is inpractical. So batch_size has to be set with respect to those hardware constraints. 

2) The size of the dataset. I.e. batch_size <= dataset_size 

3) Whether GPU or CPU is used. GPUs take advantage of many relatively weak cores to execute many operations in simultaniously. In the context of batch-size it means that if for example operations are linear-algebraic, GPU can apply operations on all data points in the batch simultaniously. Therefore, there is a point in making big batch-sizes (considering the memory constaints). On the other hand, if a code is executed on CPU, where it's possible to executute smaller number of operations in parallel, very big batch-size might not make any difference. 

4) How frequenly one wants to update parameters. Bigger batch_size implies less noise in the updates, smaller batches introduce more stochastisity into optimization.

5) GPU specification, some providers like Nvidia have a recommendation for optimal batch-sizes for specific GPUs, and one can get maximum performance by following those guidelines. 

### Mulinomial Logistic Regression: Question 3 [4 points]

Does the learning rate depend on the batch size? Explain how you should change the learning rate with respect to changes of the batch size.

Name two extreme choices of a batch size and explain their advantages and disadvantages.

**Your Answer**: It boils down to the idea that small batches(as an extreme think about 1 datapoint batches) are potentially noisy, and thus one has to be very careful about updates by setting a smaller learning rate. On the other hand, if one uses large batches (e.g. a whole dataset), he can be more confident about the updates, and thus can set higher learning rate.  

Two extremes are 1 data point batches and the whole dataset batches. The adventage of the former is that we're making more updates and in practice it works well for computationally demanding models and big datasets, as one can iterate over a dataset only once in order to train the model. On the other hand, as mentioned previously those updates are noisy, and it can cause overjumping a decent local minimum point.  The latter's advantage is that the updates are less noisy, and it 's more likely to converge to a more decent local minimum point (given that we train the model for multiple epochs). However, it most likely will be neccessary to run over the dataset multiple times in order to converge, and that is more computationally demanding than the former setup.
